```python
import math
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import time
import datetime
from statsmodels import robust
import matplotlib as mpl
from matplotlib import gridspec
import pylab
from matplotlib import cm
from pytz import timezone
```

The below graph plots the number of 2019 and 2020 false positive values as well as what those false positives are.

```python
# data to plot
n_groups = 11
n2019 = np.array([146,18,0,89,18,206,288,33,55,6,0])/184
n2020 = np.array([13,5,134,30,5,249,0,4,7,93,46])/160

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = .4
opacity = 0.8

rects1 = plt.bar(index, n2019, bar_width,
alpha=opacity/2,
color=['b'],
label='2019')

rects2 = plt.bar(index + bar_width, n2020, bar_width,
alpha=opacity,
color=['b'],
label='2020')

plt.xlabel('Events')
plt.ylabel('Events Per hour observed')
plt.title('Observed Event Rates')
plt.xticks(index + bar_width/2, ('B','S','SW','C','P','R','ST','L','NB','S','M'))
plt.legend()

#plt.tight_layout()
plt.show()
```
